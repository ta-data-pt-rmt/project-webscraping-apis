```python
import pandas as pd
import time
import re
import requests
from bs4 import BeautifulSoup
import random
import numpy as np
```

```python
headers= {'Accept-Encoding':'gzip, deflate',
          'Accept-Language':'en-US,en;q=0.9',
          'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.88 Safari/537.36'}
```

```python
cities = ['Dublin','Amsterdam','Madrid','Paris','Berlin','Milan','London','Barcelona','Frankfurt','Lisbon']
```

```python
cities_urls =[]
for city in cities:
    templist =[]
    templist.append(city)
    templist.append('https://www.numbeo.com/cost-of-living/in/'+city)
    cities_urls.append(templist)
cities_urls
```

```python
def merge_two_dicts(x, y):
    #Merge two dictionaries
    z = x.copy()   # start with keys and values of x
    z.update(y)    # modifies z with keys and values of y
    return z
```

```python
#MAIN PROGRAM ROW COUNTRY COLUMN ITEMS
cost_life_list = []
#cities_urls = ['https://www.numbeo.com/cost-of-living/in/'+city for city in cities ] 
maindf= pd.DataFrame()
#Loop to get a list of tables for cost life
count=len(cities_urls)
for citylink in cities_urls:
    time.sleep(random.randint(0,5)) # get random seconds
    count=count-1
    print("Loading....",count)
    
    
    try:
        #Validate response and get the soup
        response = requests.get(citylink[1] ,headers=headers)
        soup = BeautifulSoup(response.content)
        #print(response)
        #print(soup)
        # Lets bring the table ITEMS column soup
        table_city = soup.find_all('tr')
        
        #Temporary list to store ITEMS for each city the items 
        items_list = [] 
        
        #loop the soup to get the items ans store it in the temp list 
        for element in table_city[1:66]:
            if element.td is not None and element.td != "":
                items_list.append({'Country':citylink[0],'Item':element.td.text})
        
        
        # Lets bring the table COST column soup
        table_country = soup.find_all('span', {'class':'first_currency'})
        
        #Temporary list to store COST for each city the items 
        cost_item_list = []
        
        #loop the soup to get the COST and store it in the temp list 
        for element in table_country:
            cost_item_list.append({'Cost':element.text.replace('\xa0','').replace('€','').replace('£','')})
        
        #Merging 2 list of dictionaries with function merge_two_dicts
        table_costlife = []
       
        for element in range(len(items_list)):
            table_costlife.append(merge_two_dicts(items_list[element],cost_item_list[element]))
            
              
        #Add each city list into another list                
        cost_life_list.append(table_costlife)
        
        
        
        table_costlife
      
    except:
              
        
        pass
    #DataFrame creation
    df_table_costlife = pd.DataFrame(table_costlife)
    #Transposing the rox and columns    
    df_temp = dict(zip(df_table_costlife["Item"], df_table_costlife["Cost"]))
    df_temp = pd.DataFrame([df_temp], columns=df_temp.keys())
    df_temp['Country'] = citylink[0]
    
    
    
    maindf = pd.concat([maindf,df_temp])
    
#cost_life_list            
#print(len(cost_life_list))   
maindf

```

```python
#Main Dataframe for SUMMARY TABLE
#cost_life_summary_list = []
summary_table_cities = []
summary_table_columns = []
summary_table_costs = []
df_main_summary =  pd.DataFrame()
count=len(cities_urls)
#Loop to get a list of tables for summary
for citylink in cities_urls:
    time.sleep(random.randint(0,5)) # get random seconds
    count=count-1
    print("Loading....",count)
    try:
        #Validate response and get the soup
        response = requests.get(citylink[1] ,headers=headers)
        soup = BeautifulSoup(response.content)
        
        # Lets bring the table SUMMARY COSTS
        table_country = soup.find_all('span', {'class':'emp_number'})
        table_country = table_country[0:2]
         
        #Lets bring the table SUMMARY DETAIL for 2 in the column soup
        summary_table_text = soup.find_all('ul')
        a  = summary_table_text[6] 
        table =a.find_all('li')    
        
        #Temporary list to store costs in span class
        summary_table = []
        
         
        #loop the soup to get the items ans store it in the temp list 
        for element in table[0:2]:
            
            element = re.sub(r'([^a-zA-Z\s]+?) ', '', element.text)
            element =element.replace(' are  ', ', ')
            summary_table.append(element)
        summary_table_columns.append(summary_table)    
        
            
            
         #Temporary list to store cities
        summary_table_cities.append(citylink[0])
        
        #summary_table.append(citylink[0],)
            
        #Summary as list to LOOK BY CITY
        summary_table_costtemp = []
        for element in table_country[0:2]:
            element = element.text.replace('(3,080.97£)','').replace('(890.30£)','').replace('€','')
            summary_table_costtemp.append(element)
        summary_table_costs.append(summary_table_costtemp)
        
    except:
        pass
    
df_cities = pd.DataFrame({'City':summary_table_cities})
df_costs =pd.DataFrame(summary_table_costs,columns=summary_table_columns[0] )
maindf_summary = pd.concat([df_cities,df_costs],axis=1)
maindf_summary
```
